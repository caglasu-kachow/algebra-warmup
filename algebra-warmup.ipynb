# Noktalarƒ± √áizme

Bug√ºn√ºn ilk g√∂revleri:
1. Bir nokta √ßiz
1. Birden fazla nokta √ßiz

Baƒülantƒ±sƒ±z noktalarƒ± √ßizmek i√ßin [`plt.scatter()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter.) kullanacaƒüƒ±z.

Bir√ßok arg√ºman alabilir, ancak ilk iki pozisyonel arg√ºman en √∂nemlisidir: `x` ve `y`.

## Bir Nokta √áiz

**x=5** ve **y=10** konumlarƒ±nda bir nokta √ßiz

```python
dot_coordinates = [5, 10]

plt.scatter(dot_coordinates[0], dot_coordinates[1])
plt.show()
```

## Birden Fazla Nokta √áiz

Bir listede saklanan birka√ß noktayƒ± **boyutlar (2, 13)** (yani bir listenin listesi) ile √ßiz.

ƒ∞pucu: `x` ve `y` arg√ºmanlarƒ± x'lerin ve y'lerin listeleri olabilir.

```python
dots = [
    [1, 1, 2, 2, 2, 3, 1, 3, 3, 3, 3, 0, 0],
    [2, 3, 2, 4, 1, 3, 4, 2, 0, 5, 7, 2, 3]
]

plt.scatter(dots[0], dots[1])
plt.show()
```

Bu durumda x ve y i√ßin listeleri kullandƒ±k, ancak bu NumPy 1D-dizileri veya Pandas Serileri ile de e≈üit derecede iyi √ßalƒ±≈üƒ±r.

## Bir √áizgi √áiz

**x=4, y=0** konumunda ba≈ülayan ve **x=20, y=4** konumunda biten bir √ßizgi √ßiz. Kullanman gereken fonksiyon [pyplot.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot).

`plt.plot()` sana verdiƒüi **noktalarƒ±** √ßizer ve **noktalar arasƒ±nda d√ºz bir √ßizgi** √ßizer. Yani, bir √ßizgi √ßizmek i√ßin sadece iki nokta saƒülamanƒ±z gerekir ve `plt.plot()` bunlarƒ±n arasƒ±nda √ßizgiyi √ßizecektir.

`plt.scatter()` gibi, `plt.plot()` iki arg√ºman bekler: x'lerin bir listesi ve y'lerin bir listesi.

```python
start = {"x": 4, "y": 0}
end = {"x": 20, "y":4}

plt.plot([start["x"], end["x"]], [start["y"], end["y"]])
plt.show()
```

## Fonksiyonlarƒ± √áizme

Matematiksel bir fonksiyonu √ßizmek i√ßin her zaman bu tarifi izleyeceƒüiz:

1. √áizmek istediƒüiniz x-deƒüerlerinin aralƒ±ƒüƒ±nƒ± tanƒ±mlayƒ±n
1. Bu aralƒ±k i√ßinde e≈üit aralƒ±klƒ± x-deƒüerlerinin bir listesini veya dizisini olu≈üturun
1. Matematiksel fonksiyonunuzu kullanarak bu x-deƒüerlerine kar≈üƒ±lƒ±k gelen y-deƒüerlerini hesaplayƒ±n
1. x- ve y-deƒüerlerini √ßizin

### Bir Fonksiyonu √áiz

‚ùì Aralƒ±k `[-10, 10]` i√ßinde veri noktalarƒ± olu≈ütur. Bu noktalar senin `x` noktalarƒ±ndƒ±r. A≈üaƒüƒ±daki fonksiyonu √ßizmek i√ßin bunlarƒ± kullan:

$$y = 4x + 2$$

**Y-ekseni** **-10 ile 10** arasƒ±nda tutmak i√ßin [ plt.ylim(-10, 10)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylim.html#matplotlib.pyplot.ylim) kullan

<br/>

<details>
  <summary>üí° ƒ∞pucu</summary>
  <code><a href="https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html">numpy.arange()</a></code> veya <code><a href="https://numpy.org/doc/stable/reference/generated/numpy.linspace.html">numpy.linspace()</a></code> kullanarak bir aralƒ±k i√ßinde N sayƒ± listesi olu≈üturabilirsin.

  Fark ne?
  - Her ikisi de ba≈ülangƒ±√ß ve biti≈ü deƒüeri alƒ±r
  - `np.arange()` ile aralar i√ßin adƒ±m boyutunu tanƒ±mlarsƒ±n
  - `np.linspace()` ile aralar i√ßin nokta sayƒ±sƒ±nƒ± tanƒ±mlarsƒ±n
</details>

```python
x = np.linspace(-10, 10, 5)
y = 4*x + 2

plt.plot(x, y)
plt.ylim(-10, 10)
plt.show()
```

### Daha Geli≈ümi≈ü Fonksiyonlarƒ± √áiz

‚ùì Bu 4 fonksiyonu yeni bir grafikte birlikte √ßiz, yukarƒ±da olu≈üturulan aynƒ± `x` noktalarƒ±nƒ± kullanarak, **y-ekseni** **-2 ile 6** arasƒ±nda ve **x-ekseni** **-4 ile 4** arasƒ±nda tutarak.

$$y1 = x^2$$

$$y2 = e^x$$

$$y3 = \frac{1}{1 + e^{-(-3 + 2*x)}}$$

<details>
  <summary>üí° ƒ∞pu√ßlarƒ±</summary>
    
  Bunu yapmanƒ±n birden fazla yolu vardƒ±r... birka√ß tanesini adlandƒ±rmak gerekirse:
  - Bir `for` d√∂ng√ºs√º/liste anlayƒ±≈üƒ± kullan
  - NumPy vekt√∂rel fonksiyonlarƒ±nƒ± kullan (√∂rn. `np.exp(your_list_of_x_numbers)`)
  - Python yerle≈üik [`map`](https://www.w3schools.com/python/ref_func_map.asp) fonksiyonunu kullan

  Pro ipucu: NumPy vekt√∂rel fonksiyonlarƒ±nƒ± kullan: fonksiyonu sanki matematik yazƒ±yormu≈ü gibi yaz. Sadece daha hƒ±zlƒ± deƒüil, aynƒ± zamanda okunmasƒ± da daha kolay!

  `np.exp()` gerekli olacak.

</details>

√áok a√ßƒ±lƒ± bir √ßizgi mi alƒ±yorsun? Daha k√º√ß√ºk bir adƒ±m boyutuyla dene. `plt.plot()` sadece verdiƒüin x'leri ve y'leri √ßizer ve bunlarƒ±n arasƒ±nda **d√ºz** √ßizgi √ßizer. Noktalarƒ±nƒ±z birbirinden ne kadar yakƒ±n olursa, √ßizginiz o kadar d√ºzg√ºn olur.

```python
x = np.linspace(-4, 4, 400)

y1 = x*x
y2 = np.exp(x)
y3 = 1 / (1 + np.exp(-(-3 + 2*x)))

plt.plot(x, y1, label="x^2")
plt.plot(x, y2, label="e^x")
plt.plot(x, y3, label="logistic")

plt.xlim(-4, 4)
plt.ylim(-2, 6)

plt.legend()
plt.show()
```

## Noktalarƒ± ve √áizgileri Birle≈ütir

A≈üaƒüƒ±da `sklearn` ile olu≈üturulmu≈ü **daƒüƒ±lƒ±m grafiƒüi** var, sizin yakƒ±nda √∂ƒüreneceƒüiniz bir k√ºt√ºphane! Noktalarƒ±n daƒüƒ±lƒ±mƒ±nƒ± m√ºmk√ºn olduƒüunca yakƒ±ndan takip eden **kƒ±rmƒ±zƒ± noktalƒ± d√ºz bir √ßizgi** √ßizmek istiyoruz üñåÔ∏è

Son sonucumuz ≈ü√∂yle g√∂r√ºnmelidir:  
<img src="https://wagon-public-datasets.s3.amazonaws.com/data-science-images/02-Data-Toolkit/03-Data-Visualization/020302-regression-line-example.png" alt="regression_line" width=400>

```python
from sklearn import datasets

x, y, _ = datasets.make_regression(n_samples=100, n_features=1, noise=10, coef=True, random_state=0)

plt.scatter(x,y);
```

‚ùì Karma≈üƒ±k hesaplamalara gerek yok, sadece g√∂zlerinizi kullanarak √ßizginin ne olmasƒ± gerektiƒüini tahmin et.

Yukarƒ±daki daƒüƒ±lƒ±m grafiƒüini **yakla≈üƒ±k olarak** regrese eden kƒ±rmƒ±zƒ± noktalƒ± bir √ßizgi √ßiz
- √ñnceki h√ºcreden olu≈üturulan `x` ve `y` yi yeniden kullan
- Hatƒ±rla: `plt.plot()` ile bir √ßizgi √ßizmek i√ßin iki nokta yeterlidir
- √áizginizi kƒ±rmƒ±zƒ± ve noktalƒ± yapmak i√ßin ≈üu arg√ºmanlarƒ± kullan: `color='r', linestyle='--'` <br>(`color` i√ßin diƒüer deƒüerlerle dene: `'g'` veya `'c'` veya `'m'` ve `linestyle` i√ßin: `':'` veya `'-.'`)

```python
plt.scatter(x, y)

plt.plot([-2, 2], [-100, 100], color='r', linestyle="--")

plt.show()
```

‚ùì ≈ûimdi, grafiƒüimize baƒülam ekleyeceƒüiz:
1. "Doƒürusal Regresyon" diyen bir ba≈ülƒ±k
1. Veri noktalarƒ±nƒ± ve √ßizgiyi a√ßƒ±klayan bir a√ßƒ±klama
1. X-ekseni i√ßin etiket ("giri≈ü deƒüeri") ve y-ekseni i√ßin etiket ("√ßƒ±kƒ±≈ü deƒüeri")

Kodu buraya saƒülƒ±yoruz. Sadece `# YOUR CODE HERE` yi daha √∂nce olu≈üturduƒüun grafikle deƒüi≈ütir, ancak bu sefer iki grafiƒüe de bir `label` ekle.

```python
# Set the size of the figure
fig = plt.figure(figsize=(10, 5))

# Plot the scatter plot and the accompanying red line
# This time add a `label` to the two plots
plt.scatter(x, y, label="data points")

plt.plot([-2, 2], [-100, 100], color='r', linestyle="--", label="regression line")


# Add title and legend
plt.title("Linear Regression")
plt.legend()

# Add labels for both axes
plt.xlabel("input value")
plt.ylabel("output value")

# Add annotation to the regression line
plt.annotate("Regression line", (0,0), (1,-50), arrowprops={"arrowstyle":"->"});

plt.show()
```
